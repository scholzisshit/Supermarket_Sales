# Q10: Which product line helps you generate the most income?

# Q3: Which day of the week has maximum sales?

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['weekday'] = df['Date'].dt.day_name()
```

```python
df.set_index('Date', inplace=True)
```

```python
df.head()
```

```python
plt.figure(figsize=(8, 6))
plt.title('Daily Sales by Day of the Week')
sns.countplot(x=df['weekday'])
```

# SATURDAY HAS MOST SALES AND MONDAY HAS MINIMUM SALES, SO SUPERMARKET CAN INTRODUCE OFFER ON MONDAY TO INCREASE SALES ON MONDAY
